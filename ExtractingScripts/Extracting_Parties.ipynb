```python
import pandas as pd
```

# Reading the raw data

```python
df = pd.read_json('contratacionesabiertas_bulk.json')
```

```python
df
```

## Dropping NaN values

```python
df = df.dropna()
```

## Creating a new DataFrame called parties

```python
parties = pd.DataFrame()
```

### Extracting the parties column from the main DataFrame

```python
parties['parties'] = df['parties']
```

```python
parties
```

## Checking is there any null value

```python
parties.isnull().values.any()
```

## Adding a new _id column

```python
parties['_id'] = 1
```

```python
parties
```

### Finding the way to access to the _id

```python
df['_id'][0]['$oid']
```

## Extracting each party from each entry and attaching its entry id

```python
lista_parties = []
for i in parties.index:

    aver = []
    otra = pd.DataFrame()
    for j in range(0,len(parties['parties'][i])):
        aver.append(pd.json_normalize(parties['parties'][i][j])) 
        
    otra = pd.concat(aver)
    otra['_id'] = df['_id'][i]['$oid']
    lista_parties.append(otra)
```

## Creating a new parties DataFrame with the list of parties

```python
parties_df = pd.concat(lista_parties)
```

```python
parties_df
```

## Extracting the DataFrame into a csv with ~ separator

```python
parties_df.to_csv('parties.csv', sep='~', index=False)
```
