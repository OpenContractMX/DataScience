```python
import pandas as pd
pd.__version__
```

# Reading the raw data

```python
df_sucio = pd.read_json('contratacionesabiertas_bulk.json')
```

```python
df_sucio
```

## Making a copy of the data to work with

```python
contracts_df = df_sucio.copy()
```

## Formatting the ids correctly

```python
ids = []
for i in contracts_df.index:
    ids.append(contracts_df['_id'].iloc[i]['$oid'])
contracts_df['_id'] = ids
```

```python
contracts_df
```

## Dropping the unnecesary columns and getting rid of NaN values

```python
contracts_df = contracts_df.drop(columns=['publisher', 'cycle', 'ocid', 'id', 'date', 'tag', 'initiationType', 'parties', 'buyer', 'tender', 'language', 'awards'])
contracts_df = contracts_df.dropna()
```

```python
contracts_df
```

## Extracting all the subcontracts from the each entry and attaching its entry id

```python
contracts_list = []
for i in contracts_df.index:
    entry = pd.json_normalize(contracts_df['contracts'][i])
    entry['entry_id'] = contracts_df['_id'][i]
    contracts_list.append(entry)
```

## Converting all the list of subcontracts into a DataFrame

```python
formatted_contracts = pd.concat(contracts_list)
```

```python
formatted_contracts
```

## Dropping unnecessary columns

```python
formatted_contracts = formatted_contracts.drop(columns=['awardID', 'description', 'dateSigned', 'items'])
```

```python
formatted_contracts
```

## Exporting the DataFrame into csv file with ~ separator

```python
formatted_contracts.to_csv('FormattedContracts.csv', sep='~', index=False)
```
