### Extracting amount of contracts

Remember that each entry have more than one contract, there are like subcontracts, so we are symplifyng by summing the value of the subcontracts and placing it a new column which means the total value of the entry 

```python
amount_list = []
for i in contracts_sub_df.index:
    counter = 0
    contracts_length = len(contracts_sub_df['contracts'][i])
    
    for j in range(contracts_length):
        counter += contracts_sub_df['contracts'][i][j]['value']['amount']
    
    amount_list.append(counter)
contracts_sub_df['amount'] = amount_list
```

### Extracting the first subcontract title of each entry

```python
title_list = []
for i in contracts_sub_df.index:
    contracts_length = len(contracts_sub_df['contracts'][i])
    
    title = contracts_sub_df['contracts'][i][0]['title']
    title_list.append(title)
contracts_sub_df['title'] = title_list
```

```python
start_dates_list = []
end_dates_list = []
status_list = []

for i in contracts_sub_df.index:
    
    contracts_index = len(contracts_sub_df['contracts'][i]) - 1
    contract_start_date = contracts_sub_df['contracts'][i][0]['period']['startDate']
    
    if contracts_sub_df['contracts'][i][contracts_index]['status'] == 'terminated':

        contract_end_date = contracts_sub_df['contracts'][i][contracts_index]['period']['endDate']
    else:
        contract_end_date = 'In progress'
    
    status_list.append(contracts_sub_df['contracts'][i][contracts_index]['status'])
    start_dates_list.append(contract_start_date)
    end_dates_list.append(contract_end_date)
```

```python
contracts_sub_df['startDate'] = start_dates_list
contracts_sub_df['endDate'] = end_dates_list
contracts_sub_df['status'] = status_list
```
